# Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Customizing Plots

```python
# undergraduate degrees awarded to women in a variety of fields from 1970 to 2011.
sci_women = pd.read_csv('data/percent-bachelors-degrees-women-usa.csv')
sci_women.info()
```

```python
year = sci_women['Year']
physical_sciences = sci_women['Physical Sciences']
computer_science = sci_women['Computer Science']
health = sci_women['Health Professions']
education = sci_women['Education']
```

```python
plt.plot(year, physical_sciences, color='blue')
plt.plot(year, computer_science, color='red')
```

```python
plt.plot(year, physical_sciences, color='blue', label='Physical Sciences')
plt.plot(year, computer_science, color='red', label='Computer Science')
plt.legend(loc='lower center')
plt.xlabel('Year')
plt.ylabel('Enrollment (%)')
plt.title('Undergraduate enrollment of women')

# Set the x-axis & y-axis range
# plt.xlim([1990, 2010])
# plt.ylim([0, 50])
```

### Using subplot

```python
# Create a figure with 1x2 subplot and make the left subplot active

plt.subplot(1, 2, 1)  # nrows, ncols, nsubplot

plt.plot(year, physical_sciences, color='blue')
plt.title('Physical Sciences')

plt.subplot(1, 2, 2)
plt.plot(year, computer_science, color='red')
plt.title('Computer Science')

plt.tight_layout()   # Improve the spacing between subplots and display them
```

```python
# Create a figure with 2x2 subplot layout and make the top left subplot active

plt.subplot(2, 2, 1)
plt.plot(year, physical_sciences, color='blue')
plt.title('Physical Sciences')

plt.subplot(2, 2, 2)
plt.plot(year, computer_science, color='red')
plt.title('Computer Science')

plt.subplot(2, 2, 3)
plt.plot(year, health, color='green')
plt.title('Health Professions')

plt.subplot(2, 2, 4)
plt.plot(year, education, color='yellow')
plt.title('Education')

plt.tight_layout()
```

### annotate()

```python
plt.plot(year, computer_science, color='red', label='Computer Science') 
plt.plot(year, physical_sciences, color='blue', label='Physical Sciences')
plt.legend(loc='lower center')
plt.xlabel('Year')
plt.ylabel('Enrollment (%)')
plt.title('Undergraduate enrollment of women')

# Compute the maximum enrollment of women in Computer Science
cs_max = computer_science.max()

# Calculate the year in which there was maximum enrollment of women in Computer Science:
yr_max = year[computer_science.idxmax()]

print(cs_max, yr_max)

# Add a black arrow annotation
plt.annotate('Maximum', xy=(yr_max, cs_max), xytext=(yr_max+3, cs_max+3), arrowprops=dict(facecolor='black'))
```

## 2. Plot 2D array

```python
"""
1, 2, 1
2,-1, 1
1, 0,-1
"""

x = np.array([[1,0,-1], [2,-1,1], [1,2,1]])

plt.pcolor(x, cmap='Blues')
plt.colorbar()
plt.show()
```

```python
# Generate two 1-D arrays
u = np.linspace(-2, 2, 41)   # 41 values uniformly spaced beween -2 and +2
v = np.linspace(-1, 1, 21)

print(u)
print(v)

# Generate 2-D arrays from u and v
X,Y = np.meshgrid(u, v)

# Compute Z based on X and Y
Z = np.sin(3*np.sqrt(X**2 + Y**2)) 

print('\nXYZ shape :', Z.shape)

# Display the resulting image with pcolor()
plt.pcolor(Z)
plt.show()
```

```python
# default contour map
plt.subplot(2,2,1)
plt.contour(X, Y, Z)

# contour map with 20 contours
plt.subplot(2,2,2)
plt.contour(X, Y, Z, 20, cmap='viridis')

# default filled contour map
plt.subplot(2,2,3)
plt.contourf(X, Y, Z)

# default filled contour map with 20 contours
plt.subplot(2,2,4)
plt.contourf(X, Y, Z, 20, cmap='autumn')

plt.tight_layout()
plt.show()
```

### 2-D histogram

```python
car = pd.read_csv('data/auto-mpg.csv')
car.head()
```

```python
# 2-D histogram
plt.hist2d(car.hp, car.mpg, bins=(20,20), cmap='Blues')  # range=((40, 235), (8, 48))
plt.colorbar()

plt.xlabel('Horse power [hp]')
plt.ylabel('Miles per gallon [mpg]')
plt.title('hist2d() plot')
plt.show()
```

```python
# 2-D histogram with hexagonal bins
plt.hexbin(car.hp, car.mpg, gridsize=(20,20), cmap='Blues')  # extent=(40, 235, 8, 48)
plt.colorbar()

plt.xlabel('Horse power [hp]')
plt.ylabel('Miles per gallon [mpg]')
plt.title('hist2d() plot')
plt.show()
```
