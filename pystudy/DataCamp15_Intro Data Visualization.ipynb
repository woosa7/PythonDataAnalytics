# Intro to Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Customizing Plots

```python
# undergraduate degrees awarded to women in a variety of fields from 1970 to 2011.
sci_women = pd.read_csv('data/percent-bachelors-degrees-women-usa.csv')
sci_women.info()
```

```python
year = sci_women['Year']
physical_sciences = sci_women['Physical Sciences']
computer_science = sci_women['Computer Science']
health = sci_women['Health Professions']
education = sci_women['Education']
```

```python
plt.plot(year, physical_sciences, color='blue')
plt.plot(year, computer_science, color='red')
plt.legend(loc='lower center')
plt.xlabel('Year')
plt.ylabel('Enrollment (%)')
plt.title('Undergraduate enrollment of women')
```

```python
plt.plot(year, physical_sciences, color='blue', label='Physical Sciences')
plt.plot(year, computer_science, color='red', label='Computer Science')
plt.legend(loc='lower center')
plt.xlabel('Year')
plt.ylabel('Enrollment (%)')
plt.title('Undergraduate enrollment of women')
# Set the x-axis & y-axis range
plt.xlim([1990, 2010])
plt.ylim([0, 50])
```

### Using subplot

```python
# Create a figure with 1x2 subplot and make the left subplot active

plt.subplot(1, 2, 1)  # nrows, ncols, nsubplot

plt.plot(year, physical_sciences, color='blue')
plt.title('Physical Sciences')

plt.subplot(1, 2, 2)
plt.plot(year, computer_science, color='red')
plt.title('Computer Science')

plt.tight_layout()   # Improve the spacing between subplots and display them
```

```python
# Create a figure with 2x2 subplot layout and make the top left subplot active

plt.subplot(2, 2, 1)
plt.plot(year, physical_sciences, color='blue')
plt.title('Physical Sciences')

plt.subplot(2, 2, 2)
plt.plot(year, computer_science, color='red')
plt.title('Computer Science')

plt.subplot(2, 2, 3)
plt.plot(year, health, color='green')
plt.title('Health Professions')

plt.subplot(2, 2, 4)
plt.plot(year, education, color='gold')
plt.title('Education')

plt.tight_layout()
```

### annotate()

```python
plt.plot(year, computer_science, color='red', label='Computer Science') 
plt.plot(year, physical_sciences, color='blue', label='Physical Sciences')
plt.legend(loc='lower center')
plt.xlabel('Year')
plt.ylabel('Enrollment (%)')
plt.title('Undergraduate enrollment of women')

# Compute the maximum enrollment of women in Computer Science
cs_max = computer_science.max()

# Calculate the year in which there was maximum enrollment of women in Computer Science:
yr_max = year[computer_science.idxmax()]

print(cs_max, yr_max)

# Add a black arrow annotation
plt.annotate('Maximum', xy=(yr_max, cs_max), xytext=(yr_max+3, cs_max+3), arrowprops=dict(facecolor='black'))
```

## 2. Plot 2D array

```python
"""
1, 2, 1
2,-1, 1
1, 0,-1
"""

x = np.array([[1,0,-1], [2,-1,1], [1,2,1]])

plt.pcolor(x, cmap='Blues')
plt.colorbar()
plt.show()
```

### mesh & contour map

```python
# mesh

# Generate two 1-D arrays
u = np.linspace(-2, 2, 41)   # 41 values uniformly spaced beween -2 and +2
v = np.linspace(-1, 1, 21)

print(u)
print(v)

# Generate 2-D arrays
X,Y = np.meshgrid(u, v)

# Compute Z based on X and Y
Z = np.sin(3*np.sqrt(X**2 + Y**2)) 

print('\nXYZ shape :', Z.shape)

# Display the resulting image with pcolor()
plt.pcolor(Z)
plt.show()
```

```python
# contour map

# default contour map
plt.subplot(2,2,1)
plt.contour(X, Y, Z)

# contour map with 20 contours
plt.subplot(2,2,2)
plt.contour(X, Y, Z, 20, cmap='viridis')

# default filled contour map
plt.subplot(2,2,3)
plt.contourf(X, Y, Z)

# default filled contour map with 20 contours
plt.subplot(2,2,4)
plt.contourf(X, Y, Z, 20, cmap='autumn')

plt.tight_layout()
plt.show()
```

### 2-D histogram

```python
car = pd.read_csv('data/auto-mpg.csv')
car.head()
```

```python
# 2-D histogram
plt.hist2d(car.hp, car.mpg, bins=(20,20), cmap='Blues')  # range=((40, 235), (8, 48))

plt.colorbar()
plt.clim([0, 20])   # range of colorbar

plt.xlabel('Horse power [hp]')
plt.ylabel('Miles per gallon [mpg]')
plt.title('hist2d() plot')
plt.show()
```

```python
# 2-D histogram with hexagonal bins
plt.hexbin(car.hp, car.mpg, gridsize=(20,20), cmap='Blues')  # extent=(40, 235, 8, 48)
plt.colorbar()

plt.xlabel('Horse power [hp]')
plt.ylabel('Miles per gallon [mpg]')
plt.title('hist2d() plot')
plt.show()
```

### working with images

```python
# color image

img = plt.imread('data/baby_cat.jpg')
print(img.shape)

plt.imshow(img)
plt.axis('off')   # Hide the axes
plt.show()
```

```python
# pseudocolor

img = plt.imread('data/baby_cat.jpg')

# Compute the sum of the RGB channels
intensity = img.sum(axis=2)
print(intensity)

plt.imshow(intensity, cmap='gray')
plt.colorbar()
plt.axis('off')
plt.show()
```

```python
# Extent and aspect

img = plt.imread('data/baby_cat.jpg')

plt.figure(figsize=(7,7))

plt.subplot(2,2,1)
plt.title('extent=(-1,1,-1,1),\naspect=0.5') 
plt.xticks([-1,0,1])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-1,1,-1,1), aspect=0.5)

plt.subplot(2,2,2)
plt.title('extent=(-1,1,-1,1),\naspect=1')
plt.xticks([-1,0,1])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-1,1,-1,1), aspect=1)

plt.subplot(2,2,3)
plt.title('extent=(-1,1,-1,1),\naspect=2')
plt.xticks([-1,0,1])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-1,1,-1,1), aspect=2)

plt.subplot(2,2,4)
plt.title('extent=(-2,2,-1,1),\naspect=2')
plt.xticks([-2,-1,0,1,2])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-2,2,-1,1), aspect=2)

plt.tight_layout()
plt.show()
```

```python
# Rescaling pixel intensities

image = plt.imread('data/baby_cat.jpg')

# Extract minimum and maximum values from the image
pmin, pmax = image.min(), image.max()
print("pixel intensity : %d - %d." % (pmin, pmax))

# Rescale the pixels: rescaled_image
rescaled_image = 255*(image - pmin) / (pmax - pmin)
print("rescaled pixel intensity : %.1f - %.1f." % (rescaled_image.min(), rescaled_image.max()))

# Display the original image in the top subplot
plt.subplot(121)
plt.title('original image')
plt.axis('off')
plt.imshow(image)

# Display the rescaled image in the bottom subplot
plt.subplot(122)
plt.title('rescaled image')
plt.axis('off')
plt.imshow(rescaled_image)

plt.show()
```

```python
# histogram from a grayscale image

image = plt.imread('data/baby_cat.jpg')

plt.figure(figsize=(7,3))
plt.subplot(121)
plt.title('Original image')
plt.axis('off')
plt.imshow(image, cmap='gray')

# Flatten the image into 1 dimension: pixels
pixels = image.flatten()

plt.subplot(122)
plt.xlim((0,255))
plt.title('Normalized histogram')
plt.hist(pixels, bins=64, range=(0,256), density=True, color='red', alpha=0.4)

plt.show()
```

```python
# histogram from a color image

image = plt.imread('data/baby_cat.jpg')
plt.figure(figsize=(7,5))

plt.subplot(211)
plt.title('Original image')
plt.axis('off')
plt.imshow(image)

# Extract 2-D arrays of the RGB channels: red, blue, green
red, green, blue = image[:,:,0], image[:,:,1], image[:,:,2]
red_pixels = red.flatten()
blue_pixels = blue.flatten()
green_pixels = green.flatten()

# Overlay histograms of the pixels of each color in the bottom subplot
plt.subplot(212)
plt.title('Histograms from color image')
plt.xlim((0,255))
plt.hist(red_pixels, bins=64, normed=True, color='red', alpha=0.2)
plt.hist(blue_pixels, bins=64, normed=True, color='blue', alpha=0.2)
plt.hist(green_pixels, bins=64, normed=True, color='green', alpha=0.2)
plt.show()
```

## 3. Statistical plots with Seaborn

```python
import seaborn as sns

auto = pd.read_csv('data/auto-mpg.csv')
```

### simple linear regressions

```python
sns.lmplot(x='weight', y='mpg', data=auto)
plt.show()
```

```python
# residuals of a regression
sns.residplot(x='weight', y='mpg', data=auto, color='green')
plt.show()
```

```python
# Higher-order regressions

plt.scatter(auto['weight'], auto['mpg'], label='data', marker='o')

# linear regression of order 1
sns.regplot(x='weight', y='mpg', data=auto, color='red', order=1, scatter=None, label='order 1')
# linear regression of order 2
sns.regplot(x='weight', y='mpg', data=auto, color='gold', order=2, scatter=None, label='order 2')

plt.legend(loc='upper right')
plt.show()
```

```python
# Grouping linear regressions by hue

sns.lmplot(x='weight', y='hp', data=auto, hue='origin', palette='Set1')
plt.show()
```

```python
# Grouping linear regressions by row or column

sns.lmplot(x='weight', y='hp', data=auto, hue='origin', col='origin')
plt.show()
```

### distribution of continuous variable

```python
# strip plot
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.stripplot(x='cyl', y='hp', data=auto, jitter=False)

plt.subplot(1,2,2)
sns.stripplot(x='cyl', y='hp', data=auto, size=3, jitter=True)

plt.show()
```

```python
# swarm plots
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.swarmplot(x='cyl', y='hp', data=auto)

plt.subplot(1,2,2)
sns.swarmplot(x='hp', y='cyl', data=auto, hue='origin', orient='h')

plt.show()
```

```python
# violin plots
plt.figure(figsize=(12,5))

plt.subplot(121)
sns.violinplot(x='cyl', y='hp', data=auto)

plt.subplot(122)
sns.violinplot(x='cyl', y='hp', data=auto, inner=None, color='lightgray')
sns.stripplot(x='cyl', y='hp', data=auto, jitter=True, size=2)

plt.tight_layout()
plt.show()
```

### joint distributions

```python
sns.jointplot('hp', 'mpg', auto)
plt.show()
```

```python
sns.jointplot('hp', 'mpg', auto, kind='kde')   # kind = scatter, reg, resid, kde, hex
plt.show()
```

```python
auto_sub = auto[['mpg','weight','hp','origin']]
```

```python
sns.pairplot(auto_sub)
plt.show()
```

```python
sns.pairplot(auto_sub, hue='origin', kind='reg')
plt.show()
```

```python
corr_matrix = auto.drop(['origin'], axis=1).corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix, cmap='Blues')
plt.show()
```
